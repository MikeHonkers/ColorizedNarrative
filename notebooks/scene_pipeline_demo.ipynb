```python
from pathlib import Path
import json
import pandas as pd
from tqdm import tqdm
import sys
sys.path.append("..")
from src.scene_pipeline import ScenePipeline
pd.set_option("display.max_colwidth", None)
pd.set_option("display.max_rows", None)
pd.set_option("display.width", 2000)
```

```python
DATA_DIR = Path("../data/diarized_samples")
json_paths = sorted(DATA_DIR.glob("videoplayback_20s_diarized_*.json"))
print(f"Найдено файлов: {len(json_paths)}")
```

```python
pipeline = ScenePipeline()
```

Чтение json

```python
results = {}
for path in tqdm(json_paths, desc="Обработка файлов"):
    with path.open("r", encoding="utf-8") as f:
        payload = json.load(f)
    result = pipeline.process(payload)
    results[path.name] = {
        "payload": payload,
        "result": result,
    }
print("Все файлы обработаны.")
```

Этап 1:

```python
for name, data in results.items():
    print("\n" + "="*80)
    print(f"Исходные сегменты — {name}")
    print("="*80)
    payload = data["payload"]
    df = pd.DataFrame(payload["segments"])[["id", "start", "end", "speaker", "text"]]
    display(df)
```

Этап 2:

```python
for name, data in results.items():
    print("\n" + "="*80)
    print(f"После нормализации текста — {name}")
    print("="*80)
    result = data["result"]
    segs = result["segments"]
    df = pd.DataFrame(segs)[
        ["id", "start", "end", "speaker", "text", "normalized_text"]
    ]
    display(df)
```

Этап 3:

```python
for name, data in results.items():
    print("\n" + "="*80)
    print(f"Сцены по спикерам — {name}")
    print("="*80)

    speakers = data["result"]["speakers"]

    rows = []
    for sp in speakers:
        sc = sp.get("scene", {}) or {}
        details = sc.get("details") or []
        details = ", ".join(details) if isinstance(details, list) else str(details)

        rows.append({
            "speaker": sp.get("speaker"),
            "full_text": sp.get("text"),
            "scene": sc.get("scene", ""),
            "style": sc.get("style", ""),
            "details": details,
            "visual_prompt": sp.get("visual_prompt"),
        })

    df = pd.DataFrame(rows)
    display(df)
```
